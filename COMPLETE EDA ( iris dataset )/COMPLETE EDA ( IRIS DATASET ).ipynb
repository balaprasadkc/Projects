### COUNTPLOT FOR CATEGORICAL FEATURES

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cat
```

```python
df['species'].value_counts()
```

```python
df['species'].value_counts().plot(kind='bar',
                                 title='COUNTPLOT FOR SPECIES',
                                 figsize=(8,6))
plt.show()
```

```python
for i in cat:
    df[i].value_counts().plot(kind='bar',
                             title=f'countplot for {i}')
```

### HISTOGRAM FOR CONTINOUS FEATURES

```python
con
```

```python
for i in con:
    sns.histplot(data=df, x=i, kde=True)
    plt.title(f'histogram for {i}')
    plt.show()
                         
```

### BIVARIATE ANALYSIS

###  SCATTERPLOT & HEATMAPT FOR CON VS CON FEATURES

```python
con
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width')
plt.title(f'scatterplot sepal length vs sepal width')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='sepal_length', y='petal_length')
plt.title(f'scatterplot for sepal length vs petal length')
plt.grid()
plt.show()
```

```python
df[con].corr()
```

```python
sns.heatmap(df[con].corr(), annot=True, fmt='.3f')
```

```python
sns.heatmap(df[con].corr(), annot=True, fmt='.3f', cmap='RdYlGn')
```

### BOXPLOT FOR CATEGORICAL VS CONTINOUS FEATURES

```python
con
```

```python
cat
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title(f'boxplot for species vs sepal length')
plt.grid()
plt.show()
```

```python
for i in con:
    plt.figure(figsize=(16,8))
    sns.boxplot(data=df, x='species', y=i)
    plt.title(f'boxplot for species vs {i}')
    plt.grid()
    plt.show()
```
